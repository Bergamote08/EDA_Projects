Quelles est la distribution des prfrences dclares pour chaque attribut, sparment pour les hommes et les femmes ?


```python
#Distribution des prfences dclares par les hommes et par les femmes

#Slectionner les colonnes dsires
preferences = ["iid", "id", "gender", "attr1_1", "sinc1_1", "intel1_1", "fun1_1", "amb1_1", "shar1_1"]
df_pref = df[preferences]

#Supprimer les lignes avec valeurs manquantes
df_pref_clean = df_pref[(df_pref['iid'] < 6) | (df_pref['iid'] > 9)]

#Supprimer les lignes avec de valeurs manquantes
df_pref_clean.dropna(inplace=True)

#Dfininir les colonnes de prfrences
preference_cols = ['attr1_1', 'sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1']

#Crer une nouvelle colonne 'attribute' en additionnant les valeurs des colonnes de prfrences
df_pref_clean['attribute'] = df_pref_clean[preference_cols].sum(axis=1)

#Supprimer les lignes o 'attribute' est diffrent de 100
df_pref_clean2 = df_pref_clean[df_pref_clean['attribute'] == 100]

#Dfinir les colonnes de prferences  et leurs tiquettes
labels_preferences = ['Attractiveness', 'Sincerity', 'Intelligence', 'Fun', 'Ambition', 'Shared Interests']

#Crer les graphiques
fig, axs = plt.subplots(3, 2, figsize=(15, 15))

for ax, col, label in zip(axs.flatten(), preference_cols, labels_preferences):
  sns.kdeplot(data=df_pref_clean2[df_pref_clean['gender'] == 0], x=col, ax=ax, label='Femme')
  sns.kdeplot(data=df_pref_clean2[df_pref_clean['gender'] == 1], x=col, ax=ax, label='Homme')
  ax.set_title(f'Distribution des prfrences pour{label}')
  ax.set_xlabel('Score de prfrence')
  ax.set_ylabel('Densit')
  ax.legend()

plt.tight_layout()
plt.show()
```

Observation : les graphiques montrent des prfrences dclares pour chaque attribut, sparment pour les hommes et les femmes. on peut observer des variations dans les distributions en fonction du sexe du participant. par exemple, les femmes semblent donner plus d'iimportance  l'ambition et aux intrts partags par rapport aux hommes, tandis que les hommes donnent plus d'importance  l'attractivit.

**4. Analyse Bivarie**

Quel est le taux de match en fonction de l'thnie

```python
#Nombre de participants qui ont choisi un partenaire de mme ethnie

df_ethnie = df[['match', 'race']].dropna()
ethnie_match_rates = df_ethnie.groupby('race')['match'].mean()

#Mappage des catgories d'thnie aux nims fournis
mapping_ethnie = {
    1.0: 'Afro-Americain',
    2.0: 'Caucasien',
    3.0: 'Hispanique',
    4.0: "Asiatique",
    6.0: "Autres"
}
ethnie_match_rates_clean = ethnie_match_rates.rename(index=mapping_ethnie)

#Dfinition d'une palette de couleurs pour les barres
color = ['blue', 'orange', 'green', 'red', 'purple']

#Trac du taux correspondance par race
plt.figure(figsize=(15,6))
ethnie_match_rates_clean.plot(kind='bar', color=color)
plt.title('Taux de match en fonction de thnie')
plt.xlabel('Ethnie')
plt.ylabel('Taux de match')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show
```

Observation : on remarque que le taux de match sont assez similaires.

La diffrence d'ge entre les partenaires a t-il une incidence sur le taux de match ?

```python
#Selection des colonnes martc, age et age_o (ge du partenaire)
df_age = df[['match', 'age', 'age_o']]

#Suppression des lignes avec des valeurs man,quantes
df_age_clean = df_age.dropna()

# Craqtion de la colonne "diff_age" (diffrence d'ge)
df_age_clean['diff_age'] = df_age_clean['age']- df_age_clean['age_o']

#Copie du dataframe nettoy
df_diffage = df_age_clean.copy()

#Cration de la colonne "pos_diffage" (diffrence d'ge positive)
df_diffage['pos_diffage'] = df_diffage['diff_age'].apply(lambda x: abs(x))

# trac d'un diagramme  moustace pour visualiser la distribution
plt.figure(figsize=(15, 10))
sns.boxplot(x="match", y ="pos_diffage", data=df_diffage)
plt.title('Distribution de la diffrence ge positive vs. match')
plt.xlabel('Match')
plt.ylabel('Diffrence positive ge')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show

#Calcul du coefficient de corrlation de Spearman
correlation = df_diffage['pos_diffage'].corr(df_diffage['match'], method = 'spearman')
correlation
```

**5. Analyse Multivarie**

Quels sont les attributs qui sont les plus fortement corrls avec l'obtention d'un oui pour un deuxime rendez-vous selon l'avis du partenaire ?

```python
#Cration d"un nouveau DataFrame avec les colonnes adquate
df_dec = df[['gender', 'dec_o', 'attr_o', 'sinc_o', 'intel_o','fun_o', 'amb_o', 'shar_o']]

#Dfinition des colonnes d'attributs
attributes_columns = ['attr_o', 'sinc_o', 'intel_o','fun_o', 'amb_o', 'shar_o']

#Eliminer les lignes avec des valeurs manquantes
df_dec = df_dec.dropna()

#Calcul de la correlation entre chaque attributs et 'dec_o' pour chaque genre
correlations_attributes_dec_o = {}
for column in attributes_columns:
  correlations_attributes_dec_o[column] = (
      df_dec[df_dec['gender'] == 0][[column, 'dec_o']].corr().iloc[0, 1],
      df_dec[df_dec['gender'] == 1][[column, 'dec_o']].corr().iloc[0, 1]
  )

correlations_attributes_dec_o

#Conversion des dictionnaires de correlations en DataFrame
correlations_df = pd.DataFrame(correlations_attributes_dec_o, index=['Femmes', 'Hommes']).T

#Cration d'un barplot des correlations
correlations_df.plot(kind='bar', figsize=(15, 10), alpha=0.7)
plt.title("Correlation entre note donne par le partenaire et sa dcision d'avoir un second rdv")
plt.xlabel("Attribut donne par le partenaire")
plt.ylabel("Taux de oui (dec_o = 1)")
plt.legend(title="Genre")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show
```

Observation : Comme on peut le voir que ce soit pour les hommes et les femmes, ce sont les attributs d'attractivit (attr_o), de sens de l'amusement (fun_o) qui aident le plus  dcrocher un 2e rdv. A noter que pouyr les femmes, l'attribut d'attractivit est trs fortement corrl (>50%)  l'obtention d'un 2nd rdv.

```python
correlation = df[['dec', 'attr', 'sinc', 'intel','fun', 'amb', 'shar', 'like', 'prob']].dropna().corr()
plt.figure(figsize=(20, 10))
sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Correlation entre Attributs et Match Decision")
plt.tight_layout()
plt.show()
```

Y-a-t-il un intrt  communiquer avec un partenaire o il y a un match  la suite de l'venement ?

```python
#Slection des colonnes et renommage
df_call = df[['you_call', 'them_cal', 'match', 'date_3']]

#Filtrage des lignes o 'match' est gal  1
df_callmatch = df_call[df_call['match'] == 1]

#Ajout de la colonne 'call_made' et suppression des valeurs manques
df_callmatch['call_made'] = df_callmatch['you_call'] +df_callmatch['them_cal']
df_callmatch_clean = df_callmatch.dropna()

#Calcul de correlation de Spearman entre 'call_made' et 'date3'
spearman_correlation = df_callmatch_clean['call_made'].corr(df_callmatch_clean['date_3'], method='spearman')

#cration d'un regplot avec une ligne de rgression
plt.figure(figsize=(20, 10))
sns.regplot(x='call_made', y='date_3', data=df_callmatch_clean, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title("Nuage de Point avec Ligne de Rgression entre Call Made et Date_3")
plt.xlabel("Call Made (You Call + Them Call)")
plt.ylabel("Date_3")
plt.ylim(-0.1, 1.1)
plt.grid(True)
plt.tight_layout()
plt.show()

#Affichage de la corelation de Spearman
print(f"Corrlation de Spearman entre 'call_made et 'date_3': {spearman_correlation}")
```

Observation : Ce graphique nous montre qu'il y a une forte corrlation entre le fait de communiquer avec ses match aprs l'vnement dans le but d'avoir un second rdv. Plus la comminication augmente plus les chances d'obtenir un seconde rdv augmente.
